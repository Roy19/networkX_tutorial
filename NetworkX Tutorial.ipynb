## Simple Undirected Graphs
To create and draw a graph we need to import networkx and matplotlib.pyplot libraries into our environment.

````python
import networkx as nx
import matplotlib.pyplot as plt

# create a new empty graph
G = nx.Graph()
# add some nodes in the graph G
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
# list of all nodes added
print(G.nodes())
# add edges between nodes
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)
G.add_edge(1, 6)
# list the edges
print(G.edges())
# draw the graph G
nx.draw(G, with_labels=1)
````

````python
# create a new graph
Z = nx.Graph()
Z.add_node(1)
Z.add_node(2)
Z.add_node(3)
Z.add_node(4)
Z.add_node(5)
Z.add_node(6)
Z.nodes()
Z.add_edge(1, 2)
Z.add_edge(2, 3)
Z.add_edge(2, 4)
Z.add_edge(2, 5)
Z.add_edge(2, 6)
# change the node color
nx.draw(Z, with_labels=1, node_color='b')
````

````python
# no  of vertices
print("Vertices of G= %d" %G.order())
# no of edges
print("Edges of G= %d" %G.size())
````

## Complete Graphs
The following code illustrates a few usage of complete graphs.

````python
# use standard functions of networkx
K = nx.complete_graph(25)
# draw the complete graph
nx.draw(K)
````

````python
# create a new graph
M = nx.complete_graph(70)
nx.draw(M)
````

## Graph Layouts in NetworkX
As you can see the above graphs are pretty complex to analyze visually. So we use some definite layouts of NetworkX to add a visual effect to such type of complex graph rendering.

````python
# create a layout for analyzing complex graphs
# in this case circular rendering of the previously defined graph M
pos = nx.circular_layout(M)
nx.draw(M,pos)
````

````python
# or just one command for circular layout
nx.draw_circular(M)
````

````python
# other layout
nx.draw_spring(M)
````

## Degree Sequence and ```is_graphical()``` sequence
Create a new degree sequence as a list of integers and check if the sequence is graphical or not. Algorithm used for this - Havel Hakimi Theorem.

````python
l1 = [2, 2, 2]
# havel hakimi theorem
nx.is_graphical(l1, 'hh')
````

````python
# new degree sequence
l2 = [2, 2, 2 ,1]
nx.is_graphical(l2)
````

````python
l3 = [5, 5, 5 ,5 , 2, 2 , 2]
nx.is_graphical(l3)
````

## Path Graphs
Demonstration of path graphs in NetworkX.

````python
# draw a path graph
P = nx.path_graph(5)
nx.draw(P,with_labels=1)
````

````python
# cut vertices of path graph P
list(nx.articulation_points(P))
````

## Disjoint Graphs
Create a new graph as a union of a complete graph G and a path graph H.

````python
G = nx.complete_graph(5)
H = nx.path_graph(8)
# create a disjoint graph
K = nx.disjoint_union(G,H)
nx.draw(K, with_labels=1, node_color='g')
````

````python
# check if a graph is connected
nx.is_connected(K)
````

## Star Graph
Create a star graph in NetworkX.

````python
# star graph
S = nx.star_graph(5)
nx.draw(S, with_labels=1)
````

## Digraphs
Create and visualize directed graphs in NewtorkX.

````python
D = nx.DiGraph()
D.add_node(8)
D.add_node(5)
D.add_node('IIT')
D.add_edge(5, 'IIT')
D.add_edge(8, 'IIT')
nx.draw(D, with_labels = 1, node_color='b')
D.nodes()
````

## Bull Graph
Create a bull graph in NetworkX.

````python
# bull graph
B = nx.bull_graph()
nx.draw(B, with_labels = 1)
````

````python
# create a new graph from the graphs G and H
D.add_nodes_from(B.nodes())
D.add_edges_from(B.edges())
nx.draw(D, with_labels=1)
````
