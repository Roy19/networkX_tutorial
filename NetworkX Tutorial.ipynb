## Simple Undirected Graphs
To create and draw a graph we need to import networkx and matplotlib.pyplot libraries into our environment.

````python
import networkx as nx
import matplotlib.pyplot as plt

# create a new empty graph
G = nx.Graph()
# add some nodes in the graph G
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
# list of all nodes added
print(G.nodes())
# add edges between nodes
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)
G.add_edge(1, 6)
# list the edges
print(G.edges())
# draw the graph G
nx.draw(G, with_labels=1)
````

````python
# create a new graph
Z = nx.Graph()
Z.add_node(1)
Z.add_node(2)
Z.add_node(3)
Z.add_node(4)
Z.add_node(5)
Z.add_node(6)
Z.nodes()
Z.add_edge(1, 2)
Z.add_edge(2, 3)
Z.add_edge(2, 4)
Z.add_edge(2, 5)
Z.add_edge(2, 6)
# change the node color
nx.draw(Z, with_labels=1, node_color='b')
````

````python
# no  of vertices
print("Vertices of G= %d" %G.order())
# no of edges
print("Edges of G= %d" %G.size())
````

## Adjacency Matrix
Given a NetworkX Graph object, generate its adjacency matrix and do the same vice-versa.

````python
# create a new complete graph (example)
C = nx.complete_graph(5)
nx.draw(C)
````

We would want to import numpy package inorder to support the matrix operations required.

````python
import numpy as np
# generate the adjacency matrix A
A = nx.to_numpy_matrix(C)
# check out the Adjacency Matrix
print(A)
````

Now let's do the reverse, i.e, given an adjacency matrix create a graph.

````python
# create the networkx graph
X = nx.from_numpy_matrix(A)
X
````

````python
# now visualize it
nx.draw(X)
````

````python
# generate a custom Adjacency Matrix
B = np.matrix([[0, 1, 1, 0], [1, 0, 0, 0], 
               [1, 1, 0, 1], [0, 0, 0, 0]])
B
````

Let's see the graph.

````python
K = nx.from_numpy_matrix(B)
nx.draw(K, with_labels=1)
````

As you notice that the Adjacency Matrix B is not symmetric. But still NetworkX manages to draw the graph. In such cases NetworkX converts the matrix to a symmetric one and then draws the graph.

## Randomly Generated Graphs
So far we have covered the theory part of graph theory. Now and in the following sections, let's implement some practical applications using those learned. 
Let's create and manipulate some random graphs.

````python
import networkx as nx
import matplotlib.pyplot as plt
# generate a random graph with 10 vertices and 15 edges
G = nx.gnm_random_graph(10, 15)
nx.draw(G)
````

Do this for different times, we generate a different looking graph. Now let's create a graph with an edge being drawn with a certain probability.

````python
# random graph with 12 vertices and 0.5 probality of edges
H = nx.gnp_random_graph(30, 0.5)
nx.draw(H, node_color='g')
print('Edges = %d' %H.size())
````

Now let's do an experiment. Create graphs randomly for say n times (n is a +ve integer), for 12 vertices and 0.5 probability. Then we expect on an average that we are gonna create have graphs with approx 12C2 * 0.5 = 33 edges. Let's use a loop.

````python
import numpy as np

n_edges = []
#for loop for n = 1000
for i in range(1000):
    G = nx.gnp_random_graph(12, 0.5)
    n_edges.append(G.size())
# print the no of edges array
print(n_edges)
# print the average of the array
print('Average edges obtained = %f' %np.average(n_edges))
````

Pretty close to 33 when n is large. Try for different values of n.

## Isomorphic Graphs
Let's check condition of isomorphism on 2 different graphs.

````python
# create a new graph
G1 = nx.Graph()
G1.add_nodes_from(['KA','TN','TL','AP','KL','GO'])
G1.add_edges_from([('KA','TN'),('KA','TL'),
                   ('KA','AP'),('KA','KL'),
                  ('KA','GO')])
# let's draw it
nx.draw(G1, with_labels=1,node_color='b')
````

````python
# create a new star graph
G2 = nx.star_graph(5)
nx.draw(G2, with_labels=1)
````

````python
# check if at all they are isomorphic
nx.is_isomorphic(G1,G2)
````

## Complement of a Graph
Let's learn to create a complement of a graph.

````python
# let's create a random graph
G = nx.gnm_random_graph(15, 10)
nx.draw(G, with_labels=1)
````

````python
# let's check the complement of it
GC = nx.complement(G)
nx.draw(GC)
print(nx.is_connected(GC))
````

````python
# let's try with a complete graph
H = nx.complete_graph(5)
nx.draw(H, node_color='g')
````

````python
HC = nx.complement(H)
nx.draw(HC, node_color='b')
````

Let us now visualize self complement graph, i.e, the graph whose complement is itself. These two graphs are isomorphic. A C<sub>5</sub> graph is a perfect example of such a graph.

````python
# create a cycle
CycleGraph = nx.cycle_graph(5)
nx.draw(CycleGraph)
````

````python
CycleGraphComplement = nx.complement(CycleGraph)
nx.draw(CycleGraphComplement, node_color='b')
````

````python
nx.is_isomorphic(CycleGraph,CycleGraphComplement)
````

## Bipartite Graphs
Let's try creating bipartite graphs. But bipartite graph is persent in the sublibrary networkx.algorithms. Let's import it.

````python
from networkx.algorithms import bipartite

B = nx.Graph()
# 1st set of vertices
B.add_nodes_from([1, 2, 3, 4, 5], bipartite=0)
# 2nd set of vertices
B.add_nodes_from([1, 2, 3, 4, 5], bipartite=1)
# add the edges
B.add_edges_from([(1,'a'),(1,'b'),(2,'c'),
                 (2,'b'),(3,'c'),(4,'a'),
                  (5,'b')])
nx.draw(B, with_labels=1)
````

````python
if bipartite.is_bipartite(B):
    print('Graph is bipartite.')
else:
    print('Graph isn\'t bipartite')
````

````python
# let's try another
G = nx.path_graph(5)
if bipartite.is_bipartite(G):
    print('Graph is bipartite.')
else:
    print('Graph isn\'t bipartite')
````

````python
# and another
K = nx.gnm_random_graph(15,16)
if bipartite.is_bipartite(K):
    print('G is bipartite.')
else:
    print('G isn\'t bipartite')
````
